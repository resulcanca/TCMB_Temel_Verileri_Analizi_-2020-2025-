```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
%matplotlib inline
```

```python
plt.style.use("seaborn-v0_8-whitegrid")
sns.set_palette("deep")
pd.set_option("display.float_format", lambda x: f"{x:.2f}")
```

```python
df = pd.read_csv("tcmb_aylik_veriler.csv")
```

```python
date_cols = [col for col in df.columns if "Tarih" in col or "Date" in col or "tarih" in col]
if len(date_cols) > 0:
    df[date_cols[0]] = pd.to_datetime(df[date_cols[0]], errors='coerce')
```

```python
print(" Veri boyutu:", df.shape)
print("\n📄 Sütun isimleri:\n", df.columns.tolist())

print("\n İlk 5 satır:")
display(df.head())

print("\n Eksik veri durumu:")
display(df.isna().sum())

print("\n Özet istatistikler:")
display(df.describe())
```

```python
date_col = [col for col in df.columns if "tarih" in col.lower() or "date" in col.lower()]
if len(date_col) == 0:
    raise ValueError("Tarih sütunu bulunamadı. Lütfen tarih içeren sütun adını belirtin.")
else:
    date_col = date_col[0]
    print(f"Tarih sütunu: {date_col}")
```

```python
df[date_col] = pd.to_datetime(df[date_col], errors="coerce")
```

```python
df = df.set_index(date_col).sort_index()
```

```python
print("\n Tarih aralığı:")
print(f"{df.index.min().date()} → {df.index.max().date()}")
print(f"Toplam gözlem: {len(df)}")
```

```python
missing_ratio = (df.isna().sum() / len(df)) * 100
missing_ratio = missing_ratio[missing_ratio > 0].sort_values(ascending=False)
```

```python
if len(missing_ratio) > 0:
    print("\n⚠️ Eksik verisi bulunan sütunlar (%):")
    print(missing_ratio)
else:
    print("\n✅ Eksik veri bulunmuyor.")
```

```python
df = df.fillna(method='ffill')
```

```python
print("\n Veri temizleme tamamlandı.")
print(f"Yeni shape: {df.shape}")
df.head()
```

```python
faiz_cols = [col for col in df.columns if "faiz" in col.lower()]
if len(faiz_cols) == 0:
    raise ValueError("Faiz oranı sütunu bulunamadı. Lütfen sütun adını kontrol edin.")
else:
    faiz_col = faiz_cols[0]
    print(f"Faiz oranı sütunu: {faiz_col}")
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df[faiz_col], linewidth=2)
plt.title("TCMB Politika Faiz Oranı (2020–2025)", fontsize=14, weight='bold')
plt.xlabel("Tarih")
plt.ylabel("Faiz Oranı (%)")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
df["Faiz_Aylik_Degisim"] = df[faiz_col].diff()
```

```python
plt.figure(figsize=(12,5))
plt.bar(df.index, df["Faiz_Aylik_Degisim"], color="steelblue")
plt.axhline(0, color='gray', linewidth=1)
plt.title("Politika Faiz Oranındaki Aylık Değişim", fontsize=13, weight='bold')
plt.xlabel("Tarih")
plt.ylabel("Aylık Değişim (Puan)")
plt.show()
```

```python
print("\n Faiz Oranı Özeti:")
print(df[faiz_col].describe())

print("\n En yüksek ve en düşük faiz oranı:")
print(f"En yüksek: {df[faiz_col].max()}% ({df[faiz_col].idxmax().date()})")
print(f"En düşük:  {df[faiz_col].min()}% ({df[faiz_col].idxmin().date()})")
```

```python
rezerv_col = "resmi_rezerv_varliklari"
rezerv_cols = [col for col in df.columns if "rezerv" in col.lower() or "reserve" in col.lower()]
print(f"Bulunan rezerv sütunları: {rezerv_cols}")

if len(rezerv_cols) == 0:
    raise ValueError("Rezerv verisi bulunamadı. Lütfen veri setinde 'rezerv' geçen sütun adlarını kontrol edin.")
```

```python
brut_cols = [col for col in rezerv_cols if "brüt" in col.lower() or "gross" in col.lower()]
net_cols = [col for col in rezerv_cols if "net" in col.lower()]
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df[rezerv_col], color="teal", linewidth=2)
plt.title("TCMB Resmi Rezerv Varlıkları (2020–2025)", fontsize=14, weight='bold')
plt.xlabel("Tarih")
plt.ylabel("Milyar USD")
plt.grid(alpha=0.3)
plt.show()
```

```python
df["Rezerv_Aylik_Degisim"] = df[rezerv_col].diff()
```

```python
plt.figure(figsize=(12,5))
plt.bar(df.index, df["Rezerv_Aylik_Degisim"], color="steelblue")
plt.axhline(0, color='gray', linewidth=1)
plt.title("Resmi Rezervlerdeki Aylık Değişim", fontsize=13, weight='bold')
plt.xlabel("Tarih")
plt.ylabel("Değişim (Milyar USD)")
plt.show()
```

```python
for col in rezerv_cols:
    print(f"\n {col} Özeti:")
    print(df[col].describe())
```

```python
tufe_col = "tuketici_fiyat_endeksi"
ito_col = "ito_gecinme_endeksi"
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df[tufe_col], color="darkorange", linewidth=2)
plt.title("Tüketici Fiyat Endeksi (TÜFE) - 2020–2025", fontsize=14, weight='bold')
plt.xlabel("Tarih")
plt.ylabel("Endeks Değeri (2020=100)")
plt.grid(alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df[ito_col], color="purple", linewidth=2)
plt.title("İTO Geçinme Endeksi (İstanbul) - 2020–2025", fontsize=14, weight='bold')
plt.xlabel("Tarih")
plt.ylabel("Endeks Değeri (2020=100)")
plt.grid(alpha=0.3)
plt.show()
```

```python
df["TÜFE_Yillik_Degisim_%"] = df[tufe_col].pct_change(12) * 100
df["İTO_Yillik_Degisim_%"] = df[ito_col].pct_change(12) * 100
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df["TÜFE_Yillik_Degisim_%"], color="crimson", linewidth=2)
plt.title("TÜFE Yıllık Enflasyon Oranı (%)", fontsize=14, weight='bold')
plt.xlabel("Tarih")
plt.ylabel("Yıllık Değişim (%)")
plt.grid(alpha=0.3)
plt.show()
```

```python
faiz_col = "tuketici_kredisi_faizi"  # Faiz sütunu (senin veri setinde mevcut)
plt.figure(figsize=(12,6))
plt.plot(df.index, df["TÜFE_Yillik_Degisim_%"], label="TÜFE Enflasyonu (%)", color="firebrick", linewidth=2)
plt.plot(df.index, df[faiz_col], label="Tüketici Kredisi Faizi (%)", color="navy", linewidth=2)
plt.title("Faiz Oranı ile Enflasyon İlişkisi (2020–2025)", fontsize=14, weight='bold')
plt.xlabel("Tarih")
plt.ylabel("Oran (%)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
corr = df[[faiz_col, "TÜFE_Yillik_Degisim_%"]].corr().iloc[0,1]
print(f"\n Faiz oranı ile yıllık TÜFE enflasyonu arasındaki korelasyon: {corr:.2f}")
```

```python
fiyat_col = "konut_fiyat_endeksi"
sifir_col = "toplam_sifir_konut_satisi"
ikinciel_col = "toplam_ikinciel_konut_satisi"
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df[fiyat_col], color="darkblue", linewidth=2)
plt.title("Konut Fiyat Endeksi (2020–2025)", fontsize=14, weight='bold')
plt.xlabel("Tarih")
plt.ylabel("Endeks Değeri")
plt.grid(alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df[sifir_col], label="Sıfır Konut Satışları", color="teal", linewidth=2)
plt.plot(df.index, df[ikinciel_col], label="İkinci El Konut Satışları", color="orange", linewidth=2)
plt.title("Sıfır ve İkinci El Konut Satışları (2020–2025)", fontsize=14, weight='bold')
plt.xlabel("Tarih")
plt.ylabel("Satış Adedi")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
faiz_col = "tuketici_kredisi_faizi"

fig, ax1 = plt.subplots(figsize=(12,6))
ax1.plot(df.index, df[fiyat_col], color="darkblue", linewidth=2, label="Konut Fiyat Endeksi")
ax1.set_xlabel("Tarih")
ax1.set_ylabel("Konut Fiyat Endeksi", color="darkblue")
ax2 = ax1.twinx()
ax2.plot(df.index, df[faiz_col], color="firebrick", linewidth=2, label="Faiz Oranı")
ax2.set_ylabel("Tüketici Kredisi Faizi (%)", color="firebrick")
plt.title("Konut Fiyatları ve Faiz Oranı İlişkisi", fontsize=14, weight='bold')
fig.tight_layout()
plt.grid(alpha=0.3)
plt.show()
```

```python
corr_house_faiz = df[[fiyat_col, faiz_col]].corr().iloc[0,1]
print(f" Faiz oranı ile konut fiyat endeksi arasındaki korelasyon: {corr_house_faiz:.2f}")

corr_sifir_faiz = df[[sifir_col, faiz_col]].corr().iloc[0,1]
print(f" Faiz oranı ile sıfır konut satışları arasındaki korelasyon: {corr_sifir_faiz:.2f}")

corr_ikinciel_faiz = df[[ikinciel_col, faiz_col]].corr().iloc[0,1]
print(f" Faiz oranı ile ikinci el konut satışları arasındaki korelasyon: {corr_ikinciel_faiz:.2f}")
```

```python
usd_col = "usd_try"
eur_col = "eur_try"
gbp_col = "gbp_try"
rezerv_col = "resmi_rezerv_varliklari"
```

```python
fig, ax1 = plt.subplots(figsize=(12,6))
ax1.plot(df.index, df[rezerv_col], color="teal", linewidth=2, label="Resmi Rezerv Varlıkları")
ax1.set_xlabel("Tarih")
ax1.set_ylabel("Resmi Rezerv (Milyar USD)", color="teal")

ax2 = ax1.twinx()
ax2.plot(df.index, df[usd_col], color="firebrick", linewidth=2, label="USD/TRY")
ax2.set_ylabel("USD/TL", color="firebrick")

plt.title("Resmi Rezervler ve USD Kuru İlişkisi", fontsize=14, weight='bold')
plt.grid(alpha=0.3)
fig.tight_layout()
plt.show()
```
