```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
%matplotlib inline
```

```python
plt.style.use("seaborn-v0_8-whitegrid")
sns.set_palette("deep")
pd.set_option("display.float_format", lambda x: f"{x:.2f}")
```

```python
df = pd.read_csv("tcmb_aylik_veriler.csv")
```

```python
date_cols = [col for col in df.columns if "Tarih" in col or "Date" in col or "tarih" in col]
if len(date_cols) > 0:
    df[date_cols[0]] = pd.to_datetime(df[date_cols[0]], errors='coerce')
```

```python
print(" Veri boyutu:", df.shape)
print("\nüìÑ S√ºtun isimleri:\n", df.columns.tolist())

print("\n ƒ∞lk 5 satƒ±r:")
display(df.head())

print("\n Eksik veri durumu:")
display(df.isna().sum())

print("\n √ñzet istatistikler:")
display(df.describe())
```

```python
date_col = [col for col in df.columns if "tarih" in col.lower() or "date" in col.lower()]
if len(date_col) == 0:
    raise ValueError("Tarih s√ºtunu bulunamadƒ±. L√ºtfen tarih i√ßeren s√ºtun adƒ±nƒ± belirtin.")
else:
    date_col = date_col[0]
    print(f"Tarih s√ºtunu: {date_col}")
```

```python
df[date_col] = pd.to_datetime(df[date_col], errors="coerce")
```

```python
df = df.set_index(date_col).sort_index()
```

```python
print("\n Tarih aralƒ±ƒüƒ±:")
print(f"{df.index.min().date()} ‚Üí {df.index.max().date()}")
print(f"Toplam g√∂zlem: {len(df)}")
```

```python
missing_ratio = (df.isna().sum() / len(df)) * 100
missing_ratio = missing_ratio[missing_ratio > 0].sort_values(ascending=False)
```

```python
if len(missing_ratio) > 0:
    print("\n‚ö†Ô∏è Eksik verisi bulunan s√ºtunlar (%):")
    print(missing_ratio)
else:
    print("\n‚úÖ Eksik veri bulunmuyor.")
```

```python
df = df.fillna(method='ffill')
```

```python
print("\n Veri temizleme tamamlandƒ±.")
print(f"Yeni shape: {df.shape}")
df.head()
```

```python
faiz_cols = [col for col in df.columns if "faiz" in col.lower()]
if len(faiz_cols) == 0:
    raise ValueError("Faiz oranƒ± s√ºtunu bulunamadƒ±. L√ºtfen s√ºtun adƒ±nƒ± kontrol edin.")
else:
    faiz_col = faiz_cols[0]
    print(f"Faiz oranƒ± s√ºtunu: {faiz_col}")
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df[faiz_col], linewidth=2)
plt.title("TCMB Politika Faiz Oranƒ± (2020‚Äì2025)", fontsize=14, weight='bold')
plt.xlabel("Tarih")
plt.ylabel("Faiz Oranƒ± (%)")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
df["Faiz_Aylik_Degisim"] = df[faiz_col].diff()
```

```python
plt.figure(figsize=(12,5))
plt.bar(df.index, df["Faiz_Aylik_Degisim"], color="steelblue")
plt.axhline(0, color='gray', linewidth=1)
plt.title("Politika Faiz Oranƒ±ndaki Aylƒ±k Deƒüi≈üim", fontsize=13, weight='bold')
plt.xlabel("Tarih")
plt.ylabel("Aylƒ±k Deƒüi≈üim (Puan)")
plt.show()
```

```python
print("\n Faiz Oranƒ± √ñzeti:")
print(df[faiz_col].describe())

print("\n En y√ºksek ve en d√º≈ü√ºk faiz oranƒ±:")
print(f"En y√ºksek: {df[faiz_col].max()}% ({df[faiz_col].idxmax().date()})")
print(f"En d√º≈ü√ºk:  {df[faiz_col].min()}% ({df[faiz_col].idxmin().date()})")
```

```python
rezerv_col = "resmi_rezerv_varliklari"
rezerv_cols = [col for col in df.columns if "rezerv" in col.lower() or "reserve" in col.lower()]
print(f"Bulunan rezerv s√ºtunlarƒ±: {rezerv_cols}")

if len(rezerv_cols) == 0:
    raise ValueError("Rezerv verisi bulunamadƒ±. L√ºtfen veri setinde 'rezerv' ge√ßen s√ºtun adlarƒ±nƒ± kontrol edin.")
```

```python
brut_cols = [col for col in rezerv_cols if "br√ºt" in col.lower() or "gross" in col.lower()]
net_cols = [col for col in rezerv_cols if "net" in col.lower()]
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df[rezerv_col], color="teal", linewidth=2)
plt.title("TCMB Resmi Rezerv Varlƒ±klarƒ± (2020‚Äì2025)", fontsize=14, weight='bold')
plt.xlabel("Tarih")
plt.ylabel("Milyar USD")
plt.grid(alpha=0.3)
plt.show()
```

```python
df["Rezerv_Aylik_Degisim"] = df[rezerv_col].diff()
```

```python
plt.figure(figsize=(12,5))
plt.bar(df.index, df["Rezerv_Aylik_Degisim"], color="steelblue")
plt.axhline(0, color='gray', linewidth=1)
plt.title("Resmi Rezervlerdeki Aylƒ±k Deƒüi≈üim", fontsize=13, weight='bold')
plt.xlabel("Tarih")
plt.ylabel("Deƒüi≈üim (Milyar USD)")
plt.show()
```

```python
for col in rezerv_cols:
    print(f"\n {col} √ñzeti:")
    print(df[col].describe())
```
